**2.绘制概率密度曲线图**

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# 生成 x取值
X = np.linspace(1,50,200)

# 根据正态函数生成概率密度函数
def normal_dist(x,mean,sd):
    pro_density = (1/np.sqrt(2*np.pi*sd))*(np.exp(-1/2*((x-mean)/sd)**2))
    return pro_density

# caculate the mean and Standard deviation.
mean = np.mean(X)
sd = np.std(X)    

# application the probability density function
v = normal_dist(X,mean,sd)

# draw the figure of probability density function
plt.figure(figsize=(5,3))
plt.plot(X,v,color ='red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
plt.show()
```

**3.利用现有的包来生成概率密度曲线，并计算累积概率值**

${\color{Red} F_{X}(x) = \int_{\infty }^{x} \frac{1}{\sqrt[]{2\pi \sigma } } \exp \left [ -\frac{1}{2} \left ( \frac{x-\mu }{\sigma } \right ) ^{2} \right ]\mathrm{d}x } $

```python
# Science the normal distribution is a continue distribution, the area under the curve is the probability. 
# create a cuve of probability 

# import the basic package
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sb

# creat the distribution
data = np.arange(1,10,0.01)
pdf = norm.pdf(data, loc=5.3,scale = 1)
# the loc is nothing but the mean of data;
# the scale is the standards of deviation of data.

# visualizing the distribution
sb.set_style('whitegrid')
sb.lineplot(data,pdf,color ='red')
plt.xlabel('Heights')
plt.ylabel('Probability Density')


# 计算累积概率
# 如果我们随机选择一个人，这个人的身高小4.5英尺的概率。也就是对应上面的身高概率曲线了图中横轴小于4.5ft时，概率密度曲线与横轴所为何的面积。

#probability that the height of the person will be between 6.5 and 4.5 ft.
cdf_upper_limit = norm(loc = 5.3 , scale = 1).cdf(6.5)
cdf_lower_limit = norm(loc = 5.3 , scale = 1).cdf(4.5)
 
prob_2 = cdf_upper_limit - cdf_lower_limit
print('身高介于4.5到6.5米的概率')
print(prob_2)
 
#probability that the height of a person chosen randomly will be above 6.5ft
cdf_value = norm(loc = 5.3 , scale = 1).cdf(6.5)
prob_3 = 1- cdf_value
print('身高大于6.5米的概率')
print(prob_3)
```
